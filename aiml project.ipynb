```python
#selecting the values for X and y
X=data.iloc[:,0:-1].values
y=data.iloc[:,-1].values
```

```python
X
```

```python
y
```

```python
#splitting data into test and train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=4)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
#scaling the data to a specific range
from sklearn.preprocessing import StandardScaler
s1 = StandardScaler()
X_train=s1.fit_transform(X_train)
X_test=s1.transform(X_test)
```

# Logistic Regression

```python
#building a logistic regression prediction model
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred1=lr.predict(X_test)
```

```python
y_pred1
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred1) #calculating accuracy
```

```python
#applying decision function for AOC
y_pred_logistic = lr.decision_function(X_test)
```

```python
#shows false +ve and true +ve rate
y_pred_logistic
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred1) #calculating accuracy
```

```python
#accuracy using confusion matrix
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, ConfusionMatrixDisplay
print(classification_report(y_test,y_pred1))
```

```python
cm_logistic = confusion_matrix(y_test,y_pred1)
ConfusionMatrixDisplay(confusion_matrix=cm_logistic).plot();
```

```python
#showing roc and auc
from sklearn.metrics import roc_curve, auc

logistic_fpr, logistic_tpr, threshold = roc_curve(y_test, y_pred_logistic)
auc_logistic = auc(logistic_fpr, logistic_tpr)

plt.figure(figsize=(5, 5), dpi=100)

plt.plot(logistic_fpr, logistic_tpr, marker='.', label='Logistic (auc = %0.3f)' % auc_logistic)

plt.xlabel('False Positive Rate -->')
plt.ylabel('True Positive Rate -->')

plt.legend()

plt.show()
```

# KNearest Neighbors

```python
#using knn for fitting the model onto training set
#n_neighbors is k (k=5), 5 neighbors
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5) 
```

```python
#using c1 model to make predictions
knn.fit(X_train, y_train)
```

```python
#testing
y_pred2=knn.predict(X_test)
```

```python
y_pred2
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred2) #calculating accuracy
```

```python
#accuracy using confusion matrix
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test,y_pred2))
```

```python
cm_knn = confusion_matrix(y_test, y_pred2)
ConfusionMatrixDisplay(confusion_matrix=cm_knn).plot();
```

```python
y_score_knn = knn.predict_proba(X_test)[:, 1]

# Calculate the fpr, tpr, thresholds
fpr_knn, tpr_knn, threshold_knn = roc_curve(y_test, y_score_knn)

# Calculate the Area Under the Curve (AUC)
roc_auc_knn = auc(fpr_knn, tpr_knn)

# Plot ROC curve
plt.figure()
plt.plot(fpr_knn, tpr_knn, color='blue', lw=2, label='ROC curve (AUC = %0.2f)' % roc_auc_knn)
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ROC Curve for KNN')
plt.legend(loc='lower right')
plt.show()
```

# Ensemble Learning

```python
from sklearn.ensemble import VotingClassifier
model = VotingClassifier(estimators=[('lr', lr), ('knn', knn), ('nv',nv), ('rf', rf), ('dt', dt)], voting='hard')
model.fit(X_train,y_train)
```

```python
prediction = model.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, prediction))
```

```python
print(classification_report(y_test, prediction))
```

```python
ensemble = confusion_matrix(y_test, prediction)
ConfusionMatrixDisplay(confusion_matrix=ensemble).plot();
```

```python
#showing roc and auc

model_fpr, model_tpr, model_threshold = roc_curve(y_test,prediction)
model_auc_logistic = auc(model_fpr, model_tpr)

plt.figure(figsize=(5, 5), dpi=100)

plt.plot(model_fpr, model_tpr, marker='.', label='Logistic (auc = %0.3f)' % model_auc_logistic)

plt.xlabel('False Positive Rate -->')
plt.ylabel('True Positive Rate -->')

plt.legend()

plt.show()
```
